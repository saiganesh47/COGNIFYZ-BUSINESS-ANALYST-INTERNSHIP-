```python
import pandas as pd
from collections import Counter
df = pd.read_csv("dataset.csv")
df.head()
expectation_column = df['Expect']
expectation_lower = expectation_column.str.lower()
word_counts = Counter(expectation_lower.astype(str))
print(word_counts.most_common(15))
print(expectation_column.value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("dataset.csv")
relevant_columns = ['age', 'Duration', 'Expect']
df_select = df[relevant_columns]
df_select['Duration'] = pd.to_numeric( df_select['Duration'],errors = 'coerce')
df_select['Expect'] =pd.to_numeric( df_select['Expect'],errors = 'coerce')
correlation_matrix = df_select.corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show
sns.pairplot(df_select)
plt.show()
```
